```python
import os

a = np.array([[ -50,-40,-30,-30,-30,-30,-40,-50],
[-40,-20,  0,  0,  0,  0,-20,-40],
[-30,  0, 10, 15, 15, 10,  0,-30],
[-30,  5, 15, 20, 20, 15,  5,-30],
[-30,  0, 15, 20, 20, 15,  0,-30],
[-30,  5, 10, 15, 15, 10,  5,-30],
[-40,-20,  0,  5,  5,  0,-20,-40],
[-50,-40,-30,-30,-30,-30,-40,-50]])

# Create the directory if it doesn't exist
os.makedirs('README_images', exist_ok=True)

# Plot the heatmap
sns.heatmap(a, annot=True, cmap='RdYlGn')
plt.title("Knight piece square evaluation")
plt.savefig('README_images/heatmap.png')
plt.show()
```
